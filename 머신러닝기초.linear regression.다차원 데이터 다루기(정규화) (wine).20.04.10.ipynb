```python
import numpy as np
import matplotlib.pyplot as plt
wine=np.loadtxt('winequality-red.csv',skiprows=1,delimiter=';')
col1=0
col2=2
X=wine[:,[col1]]
y=wine[:,col2]
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
plt.scatter(X,y,c=wine[:,-1],alpha=0.3)
n=[X.min()-0.5,X.max()+0.5]
m=model.coef_*n+model.intercept_
plt.plot(n,m,'r:')
plt.colorbar()
```

```python
#     
import numpy as np
import matplotlib.pyplot as plt
wine=np.loadtxt('winequality-red.csv',skiprows=1,delimiter=';')
# x= 
X=wine[:,:-1]
# y= 
y=wine[:,-1]


from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)

w=model.coef_
b=model.intercept_

print('w=',w)
print('b=',b)
```

```python
plt.boxplot(wine)
pass
```

```python
pred_y=model.predict(X)

plt.scatter(y,pred_y,c=y)
plt.plot([y.min(),y.max()],[pred_y.min(),pred_y.max()])
#     
```

```python
X_norm=(X-X.mean(axis=0))/X.std(axis=0)
plt.boxplot(X_norm)
pass
```

```python
model=LinearRegression()
model.fit(X_norm,y)

w=model.coef_
b=model.intercept_
pred_y=model.predict(X_norm)
print(w,b)
```

```python
plt.scatter(y,pred_y,c=wine[:,-1])
plt.plot([y.min(),y.max()],[pred_y.min(),pred_y.max()])
```
