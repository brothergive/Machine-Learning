```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
X=iris.data[:,:3]
# 만약 X=iris.data[:,3] 이렇게 넣으면 차원이 맞지 않는다. ex) 에러 : x=[1,2,3] -> x=[[1],[2],[3]]
# X=iris.data[:,[2]]
# X=iris.data[:,2].reshape(-1,1)
y=iris.data[:,3]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
pred_y=model.predict(X_test)
model.score(X_test,y_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(pred_y,y_test,c=y_test)
plt.plot([0,3],[0,3],'r:')
plt.xlabel('pred_y')
plt.ylabel('test_y')
```

```python
# 원본에 대한 비율
import matplotlib.pyplot as plt
plt.scatter(pred_y,pred_y/y_test,c=y_test)
plt.xlabel('pred_y')
plt.ylabel('test_y')
plt.hlines(1,0,pred_y.max(),linestyle=':')
```

```python
index=np.argsort(y_test)
plt.plot(y_test[index],'bo:',alpha=0.3)
plt.plot(pred_y[index],'ro:',alpha=0.3)
plt.legend(['test_y','pred_y'])
```

```python
model.coef_ # u= ax+by+cz > (a,b,c) 와 비슷한 개념(가중치, 기울기)
```

```python
model.intercept_# u= ax+by+cz > (u) 와 비슷한 개념(절편)
```

```python

X=iris.data[:50,[0]]
y=iris.data[:50,1]
model=LinearRegression()
model.fit(X,y)
model.score(X,y)
```

```python

X=iris.data[50:,[0]]
y=iris.data[50:,1]
model=LinearRegression()
model.fit(X,y)
model.score(X,y)
```

```python
model.coef_, model.intercept_
```

```python
v=[X.min()-0.5,X.max()+0.5]
w=model.coef_*v+model.intercept_
# 기울기 : w(weight) ,절편 : b(ax+b)
plt.scatter(X,y)
plt.plot(v,w,'r:')
```
